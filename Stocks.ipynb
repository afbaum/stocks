```python
def CountdownBuy(stock, year, month, day):
    style.use('ggplot')
# set the start and end date for the stock data
    start = dt.datetime(year, month, day)
    end = dt.datetime(year, month *2, day)
# Get the stock data from yahoo and set the Date as an index   
    df = pd.read_csv('TSLA.csv', parse_dates=True, index_col=0)
#     df = web.DataReader(stock, 'yahoo', start, end)
    df.reset_index(inplace=True)
    df.set_index("Date", inplace=True)
    

# Shift the low column by 2 using shift method
    df['low_shift'] = df['Low'].shift(2)
    
# Shift the clsoe column by 4 using the shift method
    df['close_shift'] = df['Close'].shift(4)

# setting up calulations for buy setup and buy countdown
    df['TD_Setup'] = df['Close'] - df['close_shift']
    df['TD_Countdown'] = df['Close'] - df['low_shift']
    
# close less than four days prior
    df.loc[df.TD_Setup < 0, 'TD_Buy_Setup'] = 1
# close greater than four days prior
    df.loc[df.TD_Setup > 0, 'TD_Sell_Setup'] = 1 
    
# close less than two bar prior low
    df.loc[df.TD_Countdown < 0, 'TD_Buy_Countdown'] = 1
# close greater than two bar prior low
    df.loc[df.TD_Countdown > 0, 'TD_Sell_Countdown'] = 1 
    
# Shift the low column by 2 using shift method
    df['TD_Buy_Flip'] = df['TD_Sell_Setup'].shift(1)

# replace Nan with zero in all calculated columns
    df['TD_Buy_Setup'].fillna(0, inplace=True)
    df['TD_Sell_Setup'].fillna(0, inplace=True)
    df['TD_Buy_Countdown'].fillna(0, inplace=True)
    df['TD_Sell_Countdown'].fillna(0, inplace=True)
    df['TD_Buy_Flip'].fillna(0, inplace=True)
    
# rolling sum to look for buy and sell countdown
    df['9Buy'] = df['TD_Buy_Setup'].rolling(window=9, min_periods=0).sum()
      
# finding hits in dataframe
    beardf = df.loc[(df['TD_Buy_Flip'] == 1) & (df['TD_Buy_Setup'] == 1)]
    countdf = df.loc[(df['9Buy'] == 9)]
    
# printing dataframs
    bearFlipdf = beardf[['Close', 'close_shift', 'TD_Buy_Flip', 'TD_Buy_Setup']]
    
    
    print(countdowndf);
```
